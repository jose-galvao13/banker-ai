```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("german_credit_data.csv") # Se tiveres o csv
# Ou usa o código de simulação que te dei antes aqui

display(df.head())
print(df.info())

# "Quem pede mais crédito? Homens ou Mulheres?"
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Sex', hue='Risk')
plt.title("Risco de Crédito por Género")
plt.show()

# "A idade tem a ver com o montante?"
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de Correlação")
plt.show()
```
